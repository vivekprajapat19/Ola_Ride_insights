```python
import pandas as pd
from sqlalchemy import create_engine

# Format: 'mysql+pymysql://<username>:<password>@<host>/<dbname>'
engine = create_engine("mysql+pymysql://root:Password@localhost/ola")

# Run query and load result into DataFrame
df = pd.read_sql("SELECT * FROM cleaned_ola_data", engine)

# Show result
df.head()
```

```python
# #1 Retrieve all successful bookings:
df = pd.read_sql("SELECT * FROM cleaned_ola_data WHERE Booking_Status = 'Success' ", engine)

# Show result
df
```

```python
# 2. Find the average ride distance for each vehicle type:
df = pd.read_sql("SELECT Vehicle_Type, AVG(Ride_Distance) as avg_distance FROM cleaned_ola_data GROUP BY Vehicle_Type", engine)

# Show result
df
```

```python
# 3. Get the total number of cancelled rides by customers:
df = pd.read_sql("SELECT COUNT(*) FROM cleaned_ola_data WHERE Booking_Status = 'Canceled by Customer'", engine)

# Show result
df

```

```python
# 4. List the top 5 customers who booked the highest number of rides:

df = pd.read_sql("SELECT Customer_ID, COUNT(Booking_ID) as total_rides FROM cleaned_ola_data GROUP BY Customer_ID ORDER BY total_rides DESC LIMIT 5 ;", engine)

# Show result
df
```

```python
# 5. Get the number of rides cancelled by drivers due to personal and car-related issues:
df = pd.read_sql("SELECT COUNT(*) FROM cleaned_ola_data WHERE Canceled_Rides_by_Driver = 'Personal & Car related issue';", engine)

# Show result
df
```

```python
# 6. Find the maximum and minimum driver ratings for Prime Sedan bookings: 

df = pd.read_sql("SELECT MAX(Driver_Ratings) as max_rating, MIN(Driver_Ratings) as min_rating FROM cleaned_ola_data WHERE Vehicle_Type = 'Prime Sedan';", engine)

# Show result
df
```

```python
# 7. Retrieve all rides where payment was made using UPI:

df = pd.read_sql("SELECT * FROM cleaned_ola_data WHERE Payment_Method = 'UPI';", engine)

# Show result
df
```

```python
# 8. Find the average customer rating per vehicle type:

df = pd.read_sql("SELECT Vehicle_Type, AVG(Customer_Rating) as avg_customer_rating FROM cleaned_ola_data GROUP BY Vehicle_Type;", engine)

# Show result
df
```

```python
# 9. Calculate the total booking value of rides completed successfully:

df = pd.read_sql("SELECT SUM(Booking_Value) AS total_successful_ride_value FROM cleaned_ola_data WHERE Booking_Status = 'Success';", engine)
# Show result
df
```

```python
# 10. List all incomplete rides along with the reason
df = pd.read_sql("SELECT Booking_ID, Incomplete_Rides_Reason FROM cleaned_ola_data WHERE Incomplete_Rides = 'Yes';", engine)

# Show result
df
```
